```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns; sns.set()
```

```python
#import the salinization wetland dataset
peat_cols = ['decday', 'datetime', 'wm_gf', 'wm', 'wc_gf', 'gpp_ANNnight', 'er_ANNnight', 'wq_gf', 'TA.y', 'sal', 'PAR', 'VPD', 'PA.x', 
         'PRECIP', 'ustar', 'WT', 'H_gf', 'TW_WATER_10cm']

peat = pd.read_csv('data/peat6_all.csv', usecols=peat_cols, index_col='datetime', parse_dates=True)

#import the mature wetland (baseline) dataset
mature_cols = ['decday', 'datetime', 'wm_gf', 'wm', 'wc_gf', 'gpp_ANNnight', 'wq_gf', 'TA.y', 'sal', 'PAR', 'VPD', 'PA.x', 
         'ustar', 'WT', 'H_gf', 'TW_8cm']

mature = pd.read_csv('data/peat19_all.csv', usecols=mature_cols, index_col='datetime', parse_dates=True)
```

```python
#Fig S2: monthly fluxes showing interannual variability
monthly = peat.resample('M')['wm_gf', 'gpp_ANNnight', 'er_ANNnight', 'sal'].mean()

#unit conversions to 
monthly['wm_gf'] = monthly['wm_gf'] * 12 * 3600 * 24 * monthly.index.daysinmonth / 1E9
monthly['gpp_ANNnight'] = monthly['gpp_ANNnight'] * 12 * 3600 * 24 * monthly.index.daysinmonth / 1E6
monthly['er_ANNnight'] = monthly['er_ANNnight'] * 12 * 3600 * 24 * monthly.index.daysinmonth / 1E6

# monthly plotting function (so much more complicated than ggplot!!!)
def monthly_plotter(y, df):
    years = df.index.year.unique()
    colors = iter(cm.rainbow(np.linspace(0, 1, len(years))))

    for i in range(len(years)):
        year = df.index.year.unique()
        annual = df[df.index.year == year[i]]
        plt.plot(annual.index.month, annual[y], color=next(colors))

#This could be more concise!!!!! (Fix later)
plt.figure(figsize=(5, 14))

plt.subplot(4, 1, 1)
monthly_plotter('wm_gf', monthly)
plt.ylabel('CH4 Flux')
plt.subplot(4, 1, 2)
monthly_plotter('gpp_ANNnight', monthly)
plt.legend(year); plt.ylabel('GEP')
plt.subplot(4, 1, 3)
monthly_plotter('er_ANNnight', monthly)
plt.ylabel('ER')
plt.subplot(4, 1, 4)
monthly_plotter('sal', monthly)
plt.ylabel('Salinity'); plt.xlabel('Month')
```

```python
## Time series comparison of each site

# salinization wetland at daily timescale
peat_d = peat.resample('D')['wm_gf', 'gpp_ANNnight', 'wc_gf', 'sal'].mean()
peat_d['wm_gf'] = peat_d['wm_gf'] * 12 * 3600 * 24 / 1E6
peat_d['gpp_ANNnight'] = peat_d['gpp_ANNnight'] * 12 * 3600 *24 / 1E6
peat_d['wc_gf'] = peat_d['wc_gf'] * 12 * 3600 * 24 / 1E6

# reference wetland at daily timescale
mature_d = mature.resample('D')['wm_gf', 'gpp_ANNnight', 'wc_gf', 'sal'].mean()
mature_d['wm_gf'] = mature_d['wm_gf'] * 12 * 3600 * 24 / 1E6
mature_d['gpp_ANNnight'] = mature_d['gpp_ANNnight'] * 12 * 3600 *24 / 1E6
mature_d['wc_gf'] = mature_d['wc_gf'] * 12 * 3600 * 24 / 1E6

#plotting Fig. S4
fig, ax = plt.subplots(4, 1, figsize=(10, 6),
                      sharex='col')

# now plot it using loops!!!!!
for i in range(len(peat_d.columns)):
    ax[i].plot(peat_d[peat_d.columns[i]], linewidth = 1, label = 'Salinization wetland')
    ax[i].set_ylabel(peat_d.columns[i])
    
for i in range(len(mature_d.columns)):
    ax[i].plot(mature_d[mature_d.columns[i]], linewidth = 1, label = 'Reference wetland')
    
ax[0].legend()
```
